# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    # dp stores (min_cost, next_city_index) for path reconstruction
    dp = [[(-1, -1)] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        # Base case: All cities visited, return to the starting city (city 0)
        if mask == (1 << n) - 1:
            # Cost to return to city 0 from the current city
            # The 'next_city_index' is 0 to indicate the return to start.
            return (graph[city][0], 0)

        if dp[city][mask][0] != -1:
            return dp[city][mask]

        min_ans = float('inf')
        best_next_city = -1

        for next_city in range(n):
            if not (mask & (1 << next_city)):
                cost_from_next_city, _ = visit(next_city, mask | (1 << next_city))
                current_path_cost = graph[city][next_city] + cost_from_next_city

                if current_path_cost < min_ans:
                    min_ans = current_path_cost
                    best_next_city = next_city

        dp[city][mask] = (min_ans, best_next_city)
        return dp[city][mask]

    min_total_cost, _ = visit(0, 1)

    path = []
    current_city = 0
    current_mask = 1
    path.append(current_city)


    while current_mask < (1 << n) - 1:
        _, next_city = dp[current_city][current_mask]
        path.append(next_city)
        current_mask |= (1 << next_city) # Mark the next_city as visited
        current_city = next_city # Move to the next_city


    path.append(0)

    return min_total_cost, path
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

min_cost, path = tsp(graph)
print("Minimum travelling cost:", min_cost)
print("Optimal path:", path)
```
