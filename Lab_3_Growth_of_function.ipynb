# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
```

```python
import matplotlib.pyplot as plt
import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps

n_values = list(range(1, 1001))
time_values = []

for n in n_values:
    if n == 0:
        time_values.append(0)
        continue

    data_array = list(range(n))
    if n > 0:
        target_to_find = n - 1
        steps_taken = binary_search(data_array, target_to_find)
    else:
        steps_taken = 0

    time_values.append(steps_taken)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Binary Search Steps')
plt.plot(n_values, [math.log2(n) if n > 0 else 0 for n in n_values], linestyle='--', color='red', label='logâ(n) (Theoretical)')
plt.xlabel("Input size n")
plt.ylabel("Time (Number of steps)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.legend()
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
```

```python
import matplotlib.pyplot as plt
import math
operation_count = 0

def merge_sort(arr):
    global operation_count
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    i = j = k = 0
    while i < len(left_half) and j < len(right_half):
        operation_count += 1
        if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
        else:
            arr[k] = right_half[j]
            j += 1
        k += 1

    while i < len(left_half):
        operation_count += 1
        arr[k] = left_half[i]
        i += 1
        k += 1

    while j < len(right_half):
        operation_count += 1
        arr[k] = right_half[j]
        j += 1
        k += 1
    return arr

n_values = list(range(1, 201))
time_values = []

for n in n_values:
    if n == 0:
        time_values.append(0)
        continue

    data_list = list(range(n, 0, -1))
    operation_count = 0
    merge_sort(data_list)
    time_values.append(operation_count)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Merge Sort Operations')

scaled_log_n_values = [n * math.log2(n) if n > 0 else 0 for n in n_values]

if len(n_values) > 1 and n_values[-1] > 0:
    last_n = n_values[-1]
    last_time = time_values[-1]
    last_theoretical = last_n * math.log2(last_n)
    if last_theoretical > 0:
        scaling_factor = last_time / last_theoretical
    else:
        scaling_factor = 1
else:
    scaling_factor = 1

plt.plot(n_values, [val * scaling_factor for val in scaled_log_n_values], linestyle='--', color='red', label='n logâ(n) (Theoretical Scaled)')

plt.xlabel("Input size n")
plt.ylabel("Time (Number of Operations)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.legend()
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
```

```python
import matplotlib.pyplot as plt
import math

operation_count = 0

def heapify(arr, n, i):
    global operation_count
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2


    if left < n:
        operation_count += 1
        if arr[largest] < arr[left]:
            largest = left


    if right < n:
        operation_count += 1
        if arr[largest] < arr[right]:
            largest = right


    if largest != i:
        operation_count += 1
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    global operation_count
    n = len(arr)


    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)


    for i in range(n - 1, 0, -1):
        operation_count += 1
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

n_values = list(range(1, 201))
time_values = []

for n in n_values:
    if n == 0:
        time_values.append(0)
        continue


    data_list = list(range(n, 0, -1))
    operation_count = 0
    heap_sort(data_list)
    time_values.append(operation_count)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Heap Sort Operations')


scaled_log_n_values = [n * math.log2(n) if n > 0 else 0 for n in n_values]


if len(n_values) > 1 and n_values[-1] > 0:
    last_n = n_values[-1]
    last_time = time_values[-1]
    last_theoretical = last_n * math.log2(last_n)
    if last_theoretical > 0:
        scaling_factor = last_time / last_theoretical
    else:
        scaling_factor = 1
else:
    scaling_factor = 1

plt.plot(n_values, [val * scaling_factor for val in scaled_log_n_values], linestyle='--', color='red', label='n logâ(n) (Theoretical Scaled)')

plt.xlabel("Input size n")
plt.ylabel("Time (Number of Operations)")
plt.title("O(n log n) Log-Linear Time Complexity (Heap Sort)")
plt.legend()
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
```

```python
import matplotlib.pyplot as plt
import math
import random

operation_count = 0

def partition(arr, low, high):
    global operation_count
    pivot = arr[high]
    i = (low - 1)

    for j in range(low, high):
        operation_count += 1

        if arr[j] <= pivot:
            i = i + 1
            operation_count += 1
            arr[i], arr[j] = arr[j], arr[i]

    operation_count += 1
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return (i + 1)

def quick_sort_recursive(arr, low, high):
    global operation_count
    if low < high:

        pi = partition(arr, low, high)

        quick_sort_recursive(arr, low, pi - 1)
        quick_sort_recursive(arr, pi + 1, high)

def quick_sort(arr):
    global operation_count
    operation_count = 0
    quick_sort_recursive(arr, 0, len(arr) - 1)
    return arr

n_values = list(range(1, 201))
time_values = []

for n in n_values:
    if n == 0:
        time_values.append(0)
        continue


    data_list = [random.randint(0, 1000) for _ in range(n)]
    quick_sort(data_list)
    time_values.append(operation_count)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='Quick Sort Operations (Average Case)')


scaled_log_n_values = [n * math.log2(n) if n > 0 else 0 for n in n_values]


if len(n_values) > 1 and n_values[-1] > 0:
    last_n = n_values[-1]
    last_time = time_values[-1]
    last_theoretical = last_n * math.log2(last_n)
    if last_theoretical > 0:
        scaling_factor = last_time / last_theoretical
    else:
        scaling_factor = 1
else:
    scaling_factor = 1

plt.plot(n_values, [val * scaling_factor for val in scaled_log_n_values], linestyle='--', color='red', label='n logâ(n) (Theoretical Scaled)')

plt.xlabel("Input size n")
plt.ylabel("Time (Number of Operations)")
plt.title("O(n log n) Log-Linear Time Complexity (Quick Sort - Average Case)")
plt.legend()
plt.grid(True)
plt.show()
```
